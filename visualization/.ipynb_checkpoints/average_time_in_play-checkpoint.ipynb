# Time in play with _other statistic_

We put player's time on court in relation with his other statistcs, such as _value_, _points_ etc. This enables us to see the real **value** of the player, what he manages to do in his time, and whether the coaches shoudln't make him play more (or less).

## Defining helper functions

Find the category of a given match.

```python
def match_category(match_id):
    comp_id = matches_df.loc[match_id]['competition_id']
    return comps_df.loc[comp_id]['u_category']
```

```python
def player_matches_in_category(player_id, cats):
    player_stats = stats_df[stats_df['player_id'] == player_id]
    player_matches = matches_df.loc[player_stats['match_id'].values]
    if cats:
        mask = player_matches.index.map(lambda i: match_category(i) in cats)
        return player_matches[mask]
    else:
        return player_matches


def player_stats_in_category(player_id, cats):
    p_matches = player_matches_in_category(player_id, cats)
    mask1 = stats_df['player_id'] == player_id
    mask2 = stats_df['match_id'].apply(lambda m: m in p_matches.index)
    return stats_df[mask1 & mask2]
```

## Defining example functions

Those functions will be passed to our plotting function, which will use them to overlay the time chart with another chart. Those are the functions providing _values_, _points_ etc. They all, with the exception of `average_player_time_in_play`, have the same structure: `func(player_stats) -> int`

```python
def average_player_time_in_play(player_stats):
    time = average_time_in_play(player_stats)
    no_of_matches = len(player_stats)
    return (time, no_of_matches)


def average_time_in_play(df):
    if len(df) > 0:
        mean = df['time_in_play'].mean()
        return mean // 60
    else:
        return 0
```

```python
def average_player_value(player_stats):
    if len(player_stats) > 0:
        mean = player_stats['value'].mean()
        return round(mean, 2)
    else:
        return 0
```

```python
def average_player_points(player_stats):
    if len(player_stats) > 0:
        return player_stats['points'].mean()
    else:
        return 0
```

```python
def average_player_goals(player_stats):
    if len(player_stats) > 0:
        goals = player_stats['goals_2p'].append(player_stats['goals_3p'])
        return goals.mean()
    else:
        return 0
```

```python
def average_player_misses(player_stats):
    if len(player_stats) > 0:
        goals = player_stats['misses_2p'].append(player_stats['misses_3p'])
        return goals.mean()
    else:
        return 0
```

## Plotting the relation between time in play and _other statistic_

The following functions care of plotting the values. `plot_times_with` takes argument `data`, which is a list of dictionaries containing the information about how and with which function should the data get plotted.

```python
def generate_cat_desc(cats):
    if cats:
        return ', '.join(map( lambda c: "U" + str(c) if c != 100 else "Dosp.", cats))

    else:
        return "All categories"


def plot_times_with(data, team_id, categories= []):
    players = players_from_team(team_id)
    p_mask1, p_mask2 = players.index.duplicated(), players['name'].duplicated()
    players = players[~p_mask1 & ~p_mask2]
    player_stats = {
        p_id: player_stats_in_category(p_id, categories) for p_id in players.index
    }
    times = pd.Series(
        [average_player_time_in_play(player_stats[p]) for p in players.index],
        index=players.index,
    )
    times = times[~ times.apply( lambda x: x[1] == 0)].sort_values()
    names = [players.loc[i]['name'] for i in times.index]
    player_times = pd.Series(times.values, index=names)
    for d in data:
        func = d['func']
        stats = [func(player_stats[i]) for i in times.index]
        d['data'] = pd.Series(stats, index=names)
    time_data = {
        'title': "Average time-in-play per player",
        'data': player_times.apply( lambda x: x[0]),
        'xlabel': "Time (min)",
        'xlim': (0, 40),
        'xticks': [i for i in range(0, 41, 5)],
    }
    data.insert(0, time_data)
    with PdfPages('plots.pdf') as pdf:
        title = f"{teams_df.loc[team_id].values[0]}: {generate_cat_desc(categories)}"
        fig = get_plot(title=title, data=data, ylabel="Player names")
        pdf.savefig(fig)


def get_plot(title, data, ylabel):
    fig, axes = plt.subplots(
        nrows=1, ncols=len(data), sharey=True, figsize=(6 * len(data), 5)
    )
    fig.suptitle(title, fontsize=18)
    fig.set_dpi(250)
    for i, d in enumerate(data):
        ax = axes[i]
        d['data'].plot(kind='barh', width=0.75, ax=ax, color=['seagreen'])
        ax.set(title=d['title'], xlabel=d['xlabel'], ylabel=ylabel)
        ax.set_xlim(d['xlim'])
        ax.xaxis.set_ticks(d['xticks'])
        ax.yaxis.grid(False)
        ax.xaxis.grid(True, which='both', color='dimgray')
        ax.set_facecolor('whitesmoke')
    return fig
```

```python
points_data = {
    'title': "Player's average number of points", 
    'func': average_player_points,
    'xlabel': "Points (pts)",
    'xlim': (0, 30),
    'xticks': [i for i in range(0, 31, 3)]
}

value_data = {
    'title': "Player's average misses", 
    'func': average_player_misses,
    'xlabel': "Value (pts)",
    'xlim': (0, 15),
    'xticks': [i for i in range(0, 15, 1)]
}

plot_times_with([points_data, value_data], team_id=10020, categories=[17])
```

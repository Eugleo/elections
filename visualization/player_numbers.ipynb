# Players' number statistics

We will try to find out the distribution of the numbers of all the basketball players active in years 2015–2017.

## Theory

1. Load the player dataset.
2. Load all the players' numbers and clean the data.
3. Plot this object into a bar chart.

## Practice

### Loading the dataset

```python
import pandas as pd
```

```python
players = pd.read_csv('../players.csv', sep=';')
players.dropna(subset=['number'], inplace=True)
players['number'] = players['number'].astype('int')
players.drop_duplicates(subset=['id', 'number'], inplace=True)
players.head(3)
```

### Loading all the players' numbers

```python
from collections import Counter
```

```python
mask1 = players['number'] < 56
numbers = players[mask1]['number'].value_counts().sort_index()

mask2 = players['number'].between(55, 100)
sum_56_and_up = players[mask2]['number'].value_counts().values.sum()

values = list(map(lambda i: numbers[i] if i in numbers else 0, range(0,56)))
values.append(sum_56_and_up)

sr_numbers = pd.Series(values, index=[i for i in range(0,56)] + ['56+'])
```

### Plotting the numbers

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
ax = sr_numbers.plot(kind='bar', 
                     figsize=(15, 6),
                     title="Distribution of players' numbers",
                     width=0.75, 
                     colormap='Paired',
                     ylim=(0, 850))

ax.xaxis.set_label_text("Player number")
ax.xaxis.set_label_coords(0.5, -0.1)
ax.yaxis.set_label_text("Count of players with the number")
ax.yaxis.set_label_coords(-0.01, 0.5)

ax.tick_params(axis='y',
               which='both',
               left=False,
               right=False,
               labelleft=False)

for p in ax.patches:
    x = p.get_x() + 0.46
    
    str_num = str(p.get_height())
    if len(str_num) == 2:
        modifier = 20
    else:
        modifier = 30
    y = p.get_height() + 25 if p.get_height() < 75 else p.get_height() - modifier
    
    text = str(p.get_height())
    ax.text(x, y, text, rotation = 90, va='center', ha='center', weight='semibold')
```
